```python
import pyautogui as pya
import time
import pandas as pd
import os
import warnings
from docx2python import docx2python
import re
warnings.filterwarnings("ignore")
```

```python
def pesquisar(e):
    try:
        pya.click(pya.locateCenterOnScreen("1pesquisa.png", confidence=0.7))
    except:
        pya.click(1208,161)
    pya.write(f"{e}")
    time.sleep(1)
```

```python
def funcionando():
    while True:
        try:
            iamgem = pya.locateCenterOnScreen("1nãorespondendo.png", confidence=0.8)
            time.sleep(1)
        except:
            break
```

```python
def noaguardode(e,v):
    tempo = 0
    while True:
        try:
            imaaagem = pya.locateCenterOnScreen(e,confidence=0.8 )
            time.sleep(1)
            break
        except:
            time.sleep(1)
            tempo+=1
            if tempo==40:
                print(v)
            elif tempo >130:
                break
            pass

#noaguardode(r"C:\Users\jur.pedrorenan\Documents\automações\Suits\Imagens\buscarprocesso.png")
```

```python
resposta = pya.prompt("'.rar' ou '.docx'")
caminho_da_pasta = r"C:\Users\jur.pedrorenan\Desktop\suite"
if resposta == '.rar':
    resposta1 = pya.prompt("sua pasta já está zipada? (sim/não)")
```

```python

if resposta==".docx":
    # Defina o caminho da pasta que você deseja ler
    arquivos = os.listdir(caminho_da_pasta)

    name = [i.split(".")[0] for i in arquivos]

    name = {"Nomes": name}

    df = pd.DataFrame(name)
    df["status"]="Pendente"
    df["Observações"]=""

    df.insert(1, "processo", "")
    for arquivo in arquivos:
        caminhodoarquivo = os.path.join(caminho_da_pasta,arquivo)
        with docx2python(caminhodoarquivo) as document:
            texto = document.text
            indice = texto.find(".8.")
            processo = texto[int(indice)-15 : int(indice)+10]
        for i,v in enumerate(df["Nomes"]):
            if arquivo.split(".")[0] == v:
                df["processo"].iloc[i] = processo
elif resposta==".rar":
    if resposta1=="sim":
        arquivos = os.listdir(caminho_da_pasta)
        name = [i.split(".ra")[0] for i in arquivos]
        name = {"Processo": name}
        df = pd.DataFrame(name)
        df["status"] = "Pendente"
        df["Observações"] = ""
    else:
        name = os.listdir(caminho_da_pasta)
        name = {"Processo": name}
        df = pd.DataFrame(name)
        df["status"]="Pendente"
        df["Observações"] = ""
    
```
